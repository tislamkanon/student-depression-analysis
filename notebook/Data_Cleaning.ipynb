```python
#Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("student_depression_dataset.csv")
```

```python
df.info()
```

```python
#Display first few rows
df.head()
```

```python
#Check for Missing Values
df.isnull().sum()
```

```python
#Remove Duplicates
df = df.drop_duplicates()
```

```python
df
```

```python
#Convert Data Types
df["Age"] = df["Age"].astype(int)
df["Work/Study Hours"] = df["Work/Study Hours"].astype(int)
```

```python
df
```

```python
df["Sleep Duration"] = df["Sleep Duration"].str.replace("'", "").str.strip()
```

```python
#Handle Missing Values
df["Financial Stress"].unique()
```

```python
#Replace ? with NaN and convert to float
df["Financial Stress"] = df["Financial Stress"].replace("?", np.nan).astype(float)
```

```python
#Handle Missing Values in "Financial Stress"
df["Financial Stress"].fillna(df["Financial Stress"].mean(), inplace=True)
```

```python
df["Financial Stress"].fillna(df["Financial Stress"].median(), inplace=True)
```

```python
df.dropna(subset=["Financial Stress"], inplace=True)
```

```python
#Save Cleaned Dataset
df.to_csv("cleaned_student_depression_dataset.csv", index=False)
```
